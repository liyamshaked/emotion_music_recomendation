```python
%%writefile "/content/drive/MyDrive/Colab Notebooks/project/final/pages/base_page.py"
"""
Base page module for the music recommender app.
Provides common functionality for all pages.
"""
import sys
import importlib

# Import modules with reload to avoid circular dependencies
sys.path.append('/content/drive/MyDrive/Colab Notebooks/project/final/app')
import layout
importlib.reload(layout)

import ipywidgets as widgets
from IPython.display import display, HTML

sys.path.append('/content/drive/MyDrive/Colab Notebooks/project/final/app')
import navigation
importlib.reload(navigation)
from navigation import get_output

class BasePage:
    """Base class for all pages."""

    def __init__(self, title="Music Recommender"):
        """
        Initialize a new page.

        Args:
            title (str): The page title
        """
        self.title = title
        self.output = get_output()
        self.widgets = []

    def create_title(self):
        """Create the page title."""
        return layout.create_title(self.title)

    def create_container(self):
        """Create a container with all page widgets."""
        return layout.create_page_container(self.widgets)

    def render(self):
        """Render the page (to be implemented by subclasses)."""
        raise NotImplementedError("Subclasses must implement render()")

    def display(self):
        """Display the page."""
        container = self.create_container()
        display(container)

    def add_widget(self, widget):
        """Add a widget to the page."""
        self.widgets.append(widget)

    def add_spacer(self, height=10):
        """Add a vertical spacer to the page."""
        self.add_widget(layout.create_spacer(height))

    def add_title(self):
        """Add the page title to the widgets list."""
        self.add_widget(self.create_title())

    def add_subtitle(self, text):
        """Add a subtitle to the page."""
        self.add_widget(layout.create_subtitle(text))

    def add_button(self, description, button_style='', icon='', on_click=None):
        """Add a button to the page."""
        button = layout.create_button(description, button_style, icon, on_click)
        self.add_widget(button)
        return button

    def add_buttons_container(self, buttons, justify='center'):
        """Add a container with buttons to the page."""
        container = layout.create_buttons_container(buttons, justify)
        self.add_widget(container)

    def add_form_field(self, field_type, description, **kwargs):
        """Add a form field to the page."""
        field = layout.create_form_field(field_type, description, **kwargs)
        self.add_widget(field)
        return field
```
