```python
import pandas as pd
import os

project_dir = '/content/drive/MyDrive/Colab Notebooks/project/Music/skeleton/data/'

# ---------- Create users.csv ----------
users_df = pd.DataFrame({
    'user_id': [1, 2],
    'first_name': ['Alice', 'Bob'],
    'last_name': ['Smith', 'Jones'],
    'age': [25, 30],
    'gender': ['F', 'M']
})
users_df.to_csv(os.path.join(project_dir, 'users.csv'), index=False)

# ---------- Create emotions.csv ----------
emotions_df = pd.DataFrame({
    'emotion_id': [1, 2, 3, 4],
    'emotion_str': ['happy', 'sad', 'angry', 'relaxed'],
    'positivity': [1.0, -1.0, -0.8, 0.7]
})
emotions_df.to_csv(os.path.join(project_dir, 'emotions.csv'), index=False)

# ---------- Create listening_history.csv ----------
listening_history_df = pd.DataFrame(columns=[
    'user_id', 'song_id', 'mood_before', 'mood_after',
    'start_date', 'end_date', 'is_recommended'
])
listening_history_df.to_csv(os.path.join(project_dir, 'listening_history.csv'), index=False)

print("✅ CSV files saved to:", project_dir)
```
