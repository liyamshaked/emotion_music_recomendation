```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import os

# Load main song dataset
file_path = "/content/drive/MyDrive/Colab Notebooks/project/Music/cleaned_df.csv"
df = pd.read_csv(file_path)

# Prepare users and emotions
users = [f"user_{i}" for i in range(1, 101)]
emotions = df['emotion'].unique()
n_sessions_per_user = 100

# Prepare save directory
save_dir = "/content/drive/MyDrive/Colab Notebooks/project/Music/users history"
os.makedirs(save_dir, exist_ok=True)

# Generate per-user files
for user in users:
    records = []
    start_time = datetime.now() - timedelta(days=30)  # simulate last month

    for _ in range(n_sessions_per_user):
        song = df.sample(1).iloc[0]
        song_id = song['song']
        artist = song['artist']
        song_emotion = song['emotion']
        mood_before = random.choice(emotions)

        # Simulate mood change
        if mood_before in ["sadness", "anger"] and song_emotion in ["joy", "love"]:
            mood_after = "joy"
        else:
            mood_after = mood_before

        mood_improved = int(mood_after != mood_before)
        timestamp = start_time + timedelta(minutes=random.randint(0, 60 * 24 * 30))

        record = {
            "user_id": user,
            "song_id": song_id,
            "artist": artist,
            "mood_before": mood_before,
            "mood_after": mood_after,
            "mood_improved": mood_improved,
            "timestamp": timestamp.strftime("%Y-%m-%d %H:%M:%S")
        }
        records.append(record)

    # Create DataFrame and save per user (inside the loop!)
    user_df = pd.DataFrame(records)
    user_df.to_csv(f"{save_dir}/{user}_history.csv", index=False)
    print(f"✅ Saved {save_dir}/{user}_history.csv with {len(records)} records")
```
